## Let's do something. let's check app wise. How Vader Performs. 

```python
def accuracy_vader(app_name):
    df_new = df[df['app']==app_name]
    df_new['pred_score'] = df_new['content'].apply(lambda x: sid.polarity_scores(x)['compound'])
    df_new['pred_ratings'] = df_new['pred_score'].apply(lambda x : "pos" if x>0 else "neg" if x<0 else "neu")
    correct_predictions = 0
    for k in range(df_new.shape[0]):
        if df_new.iloc[k]['review_ratings'] == df_new.iloc[k]['pred_ratings']:
            correct_predictions += 1
    return correct_predictions / df_new.shape[0]
```

```python
df['app'].value_counts().index
```

```python
for k in df['app'].value_counts().index : 
    print(f"Accuracy for app - {k} is - {accuracy_vader(k)}")
```
