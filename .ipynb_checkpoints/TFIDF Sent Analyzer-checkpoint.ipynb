```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('L:\\Workspace\\Models\\Review Sent Analyzer And Summarizer\\cleaned_reviews_data_20_apps.csv')
df
```

```python
df_new = df[df['review_ratings']!='neu']
df_new.shape
```

```python
X = df_new['content']
y = df_new['review_ratings']
print(" X:", X.shape , " Y: ", y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
x_train_vec = tfidf.fit_transform(x_train)
x_test_vec = tfidf.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(x_train_vec, y_train)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC()
svc.fit(x_train_vec, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_vec, y_train)
```

```python
from sklearn.naive_bayes import ComplementNB

cnb = ComplementNB()
cnb.fit(x_train_vec, y_train)
```

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(loss="log_loss") 
sgd.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
def print_scores(model):
    preds = model.predict(x_test_vec)
    acc = accuracy_score(preds, y_test)
    print(f"Accuracy = {acc}")
    print("Confusion Matrix:")
    print(confusion_matrix(preds, y_test))
    print("\n Classification Report - ")
    print(classification_report(preds, y_test))
```

```python
print("MultinomialNB - ")
print_scores(nb)
```

```python
print("LinearSVC - ")
print_scores(svc)
```

```python
print("LogisticRegression - ")
print_scores(lr)
```

```python
print("ComplementNB - ")
print_scores(cnb)
```

```python
print("SGD Classfier - ")
print_scores(sgd)
```

## Now lets try with neural as well? 

```python
X_new = df['content']
y_new = df['review_ratings']
```

```python
x_train_new, x_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.33, random_state=42)
```

```python
tfidf_new = TfidfVectorizer()
x_train_new_vec = tfidf_new.fit_transform(x_train_new)
x_test_new_vec = tfidf_new.transform(x_test_new)
```

```python
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC

def print_scores_new():
    models = {
        "nb": MultinomialNB(),
        "cnb": ComplementNB(),
        "logreg": LogisticRegression(max_iter=1000),
        "svm": LinearSVC(),
        "sgd": SGDClassifier()
    }
    for k in models.keys():
        model = models[k]
        model.fit(x_train_new_vec, y_train_new)
        print("Model - ", k)
        preds = model.predict(x_test_new_vec)
        acc = accuracy_score(y_test_new,preds)
        print(f"Accuracy = {acc}")
        print("Confusion Matrix:")
        print(confusion_matrix(y_test_new,preds))
        print("\n Classification Report - ")
        print(classification_report(y_test_new,preds))
```

```python
print_scores_new()
```
