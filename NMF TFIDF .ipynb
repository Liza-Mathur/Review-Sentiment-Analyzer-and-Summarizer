```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('L:\\Workspace\\Models\\Review Sent Analyzer And Summarizer\\cleaned_reviews_data_20_apps.csv')
```

```python
df.head()
```

```python
df_neg = df[df['review_ratings']=='neg'][['content']]
df_neg.shape
```

```python
df_neg.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df = 0.95 , min_df = 2, stop_words = 'english')
```

```python
content_vecs = tfidf.fit_transform(df_neg['content'])
content_vecs.shape
```

```python
from sklearn.decomposition import NMF 
```

```python
nmf = NMF(n_components=6, random_state=42)
nmf.fit(content_vecs)
```

```python
for index, topic in enumerate(nmf.components_):
    print(f"The top 30 words for topic numner {index} : ")
    print([tfidf.get_feature_names_out()[index] for index in topic.argsort()[-30:]])
    print('\n')
```

```python
topics_res = nmf.transform(content_vecs)
```

```python
df_neg['Topic'] = topics_res.argmax(axis=1)
```
