```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('L:\\Workspace\\Models\\Review Sent Analyzer And Summarizer\\cleaned_reviews_data_20_apps.csv')
```

```python
df['review_ratings'] = df['review_ratings'].replace('neu', 'pos')
df['review_ratings'].value_counts()
```

```python
X = df['content']
y = df['review_ratings']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
tfidf = TfidfVectorizer()
x_train_vec = tfidf.fit_transform(x_train)
x_test_vec = tfidf.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def print_scores_new():
    models = {
        "nb": MultinomialNB(),
        "cnb": ComplementNB(),
        "logreg": LogisticRegression(max_iter=1000),
        "svm": LinearSVC(),
        "sgd": SGDClassifier()
    }
    for k in models.keys():
        model = models[k]
        model.fit(x_train_vec, y_train)
        print("Model - ", k)
        preds = model.predict(x_test_vec)
        acc = accuracy_score(y_test,preds)
        print(f"Accuracy = {acc}")
        print("Confusion Matrix:")
        print(confusion_matrix(y_test,preds))
        print("\n Classification Report - ")
        print(classification_report(y_test,preds))
```

```python
print_scores_new()
```
